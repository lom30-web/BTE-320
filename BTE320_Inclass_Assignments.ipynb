Recursive function: string reversal

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return s[-1] + reverse(s[:-1]) #Take the last character first
reverse('hello')
```

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return reverse(s[1:]) + s[0] #Take the first character last
reverse('12345')
```

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return s[-1] + reverse(s[1:-1]) + s[0] #Swap first and last characters
reverse('12345')
```
