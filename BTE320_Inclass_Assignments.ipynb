```python
Sales = float(input("Enter Sales ($) : "))
TaxRate= float(input("Enter the Tax Rate (%) : "))
CostofGoodsSold = 40/100 * Sales
GrossProfit = Sales - CostofGoodsSold
Tax = TaxRate/100 * GrossProfit
NetIncome = GrossProfit - Tax
print(f"Sales : $ {Sales}")
print(f"Cost of Goods Sold : $ {CostofGoodsSold}")
print(f"Gross Profit : $ {GrossProfit}")
print(f"Tax : $ {Tax}")
print(f"Net Income : $ {NetIncome}")
```

WEEK 3 - Conditional Statements

```python
a = float (input ("Enter first number: "))
b = float (input ("Enter second number: "))
MathOperator = str (input ("Enter mathematical operator of choice: "))
if MathOperator == "+":
  output = a + b
  print (output)
elif MathOperator == "-":
  output = a - b
  print (output)
elif MathOperator == "*":
  output = a * b
  print (output)
elif MathOperator == "/":
  output = a / b
  print (output)
elif MathOperator == "**":
  output = a ** b
  print (output)
elif MathOperator == "%":
  output = a % b
  print (output)
elif MathOperator == "//":
  output = a // b
  print (output)
else:
  print ("Invalid mathematical operator")
```

# **IN CLASS WALKTHROUGH**

```python
num1 = float (input ("Enter first number: "))
num2 = float (input ("Enter second number: "))
MathOperator = str (input ("Enter mathematical operator of choice: "))
result = None
if MathOperator == "+":
  result = num1 + num2
elif MathOperator == "-":
  result = num1 - num2
elif MathOperator == "*":
  result = num1 * num2
elif MathOperator == "/":
  result = num1 / num2
elif MathOperator == "**":
  result = num1 ** num2
elif MathOperator == "%":
  result = num1 % num2
elif MathOperator == "//":
  result = num1 // num2
else:
  print ("Invalid mathematical operator")
print(f"Result: {result}")
```
