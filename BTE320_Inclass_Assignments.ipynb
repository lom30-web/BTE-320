# Recursive function: string reversal

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return s[-1] + reverse(s[:-1]) #Take the last character first
reverse('hello')
```

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return reverse(s[1:]) + s[0] #Take the first character last
reverse('12345')
```

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return s[-1] + reverse(s[1:-1]) + s[0] #Swap first and last characters
reverse('12345')
```

# Loan Payment Calculator

```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float)) and isinstance(annual_rate, (float)) and isinstance(months, (int,float))
  try:
    monthly_rate = annual_rate / 12
    if monthly_rate != 0:
      monthly_payment = monthly_rate * principal / (1 - (1 + monthly_rate) ** (-months))
    else:
     monthly_payment = principal / months
  except:
    raise ValueError("Invalid input; Function inputs must be positive")
  return monthly_payment

principal = float(input("Enter the loan amount: "))
annual_rate = float(input("Enter the annual interest rate: "))
months = float(input("Enter the number of months: "))
print (round(loan_payment(principal, annual_rate, months)))
```

```python
def getValues(n):
  values = []
  for i in range(n):
    while True:
      try:
        x = float(input(f"Enter value {i+1}: "))
        if x < 0:
          raise ValueError
        values.append(x)
        break
      except:
        print("Invalid input. Enter a non-negative number.")
  return values

def calcAverage(alist):
  valid_values = alist[3:]
  if len(valid_values) == 0:
    return 0

  return sum(valid_values) / len(valid_values)

n = int(input("How many values? "))
data = getValues(n)
average = calcAverage(data)

print("Average (excluding first 3 values):", average)
```

# Student Class

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    assert isinstance(newName, str)
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print (f"Student {self.__name} is already enrolled in {newCourse}")
```

```python
s = Student("Letsile Mogwe", 12345)
print(f"Student name: {s.get_name()}")
print(f"Student number: {s.number}")
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll("BTE 320")
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll("BTE 320")
```

# Classes Point and Rectangle

```python
import math

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle:
  def __init__(self, p1, p2, p3, p4):
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4

  def width(self):
    return self.p1.distance(self.p2)

  def height(self):
    return self.p1.distance(self.p4)

  def area(self):
    return self.width() * self.height()
```

```python
p1 = Point(0, 0)
p2 = Point(4, 0)
p3 = Point(4, 3)
p4 = Point(0, 3)

rect = Rectangle(p1, p2, p3, p4)

print("Width:", rect.width())
print("Height:", rect.height())
print("Area:", rect.area())
```

# Class Employee - Manager

```python
class Employee:
 def __init__(self, name, hourly_wage, hours_worked):
  self.name = name
  self._hourly_wage = hourly_wage
  self.hours_worked = hours_worked

 def calculate_pay(self):
  return self._hourly_wage * self.hours_worked

 def display_info(self):
  print(f"Name: {self.name}")
  print(f"Hourly Wage: ${self._hourly_wage}")
  print(f"Hours Worked: {self.hours_worked} hours")

 def get_wage(self):
  return self._hourly_wage

 def set_wage(self, new_wage):
  if new_wage <= 0:
    print("Invalid wage value: Wage cannot be negative.")
  else:
    self._wage = new_wage

class Manager(Employee):
 def __init__(self, name, hourly_wage, hours_worked, bonus):
  super().__init__(name, hourly_wage, hours_worked)
  self.bonus = bonus #decimal format, e.g, 0.2 for 20%

 def calculate_pay(self):
  return (1 + self.bonus) * self.get_wage() * self.hours_worked
```

```python
employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)

print("Employee Info:")
employee.display_info()
print(f"Weekly Pay: ${employee.calculate_pay()}")

print("\nManager Info:")
manager.display_info()
print(f"Weekly Pay: ${manager.calculate_pay()} which includes a manager bonus of {manager.bonus * 100}%")
```
